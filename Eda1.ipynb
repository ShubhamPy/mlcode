```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import numpy as np
```

```python
rd_csv=pd.read_csv("kc_house_data.csv")
hpr=pd.DataFrame(rd_csv)
hpr.head()
```

```python
hpr.dtypes
```

```python
hpr.info()
```

```python
sns.kdeplot(hpr['price'],shade=True,bw=0.05)
```

```python
sns.kdeplot(hpr['long'],shade=True)
```

```python
hpr['bedrooms'].value_counts().plot(kind='bar')
plt.title('number of Bedroom')
plt.xlabel('Bedrooms')
plt.ylabel('Count')
#sns.despine
```

```python
hpr['floors'].value_counts().plot(kind='bar')
plt.title('number of floors')
plt.xlabel('Floors')
plt.ylabel('Count')
sns.despine
```

```python
sns.jointplot(x='lat',y='long',data=hpr,size=10)
sns.despine
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x="price",y='sqft_living',data=hpr)
```
